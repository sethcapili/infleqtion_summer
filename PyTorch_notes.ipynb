## Seeing models as computation graphs
--- 
- we covered Pytorch's tensor library but now we need to look into pytorch's automatic differentiation system 
- the module base class makes it easier to train models. 
- it allows us to encapsulate layers and operations and keep track of the model’s parameters

- computational graphs let us vizualize math concepts, in deep learning a computation graphn lets us veiw the math needed to output a neural network
- we need the calculations to compute the gradients for backpropogation which is the primary training algorithm for neural networks

```python
import torch.nn.functional as F

y = torch.tensor([1.0]) # true label
x1 = torch.tensor([1.1]) # input feature
w1 = torch.tensor([2.2]) # weight
b = torch.tensor([0.0]) # bias

z = x1 * w1 + b # net input 
a = torch.sigmoid(z) # activation function
loss = F.binary_cross_entropy(a, y) # calculate the loss
print(loss)
```

The point of this logistic regression classifier isnt to implement it but more so help us understand how we can think of a sequance of computations as a computation graph

- pytorch builds compuation graphs in the background and we can use this to calculate the gradient of a loss function that follows the weight and bias models to train the model. 

## Automatic differentiation 
---
- gradients are required in training nn with backpropogation
- the chain rule is important as it is how nn calculate their errors within each layer and passes the error backward (backpropogation) to correct said errors 
- partial derivatives "freeze" variables and tweak one at the time to create a more structured tuning. (cake idea w/ deliciousness)
- "When PyTorch wants to update $w_1$, it uses a partial derivative ($\frac{\partial L}{\partial w_1}$). It asks: "If I hold the bias $b$ completely still, how much will tweaking $w_1$ reduce my error?"By calculating the partial derivative for $w_1$ and $b$ separately, PyTorch knows exactly how much to turn each individual knob to make the AI smarter."
- gradients are vectors that contain all the partial derivatives of a functoon with muliple variables as inputs. 

- intuitively, the chain rule is used to compute the loss function of a model's parameters in a cg. Loss function works as a tool for mesuring a models's performance

- the autograd engine creates a cg in the background. when we call the grad function, we compute the gradient of loss to the models parameter weight (w1). 



```python
import torch
import torch.nn.functional as F
from torch.autograd import grad

y = torch.tensor([1.0])
x1 = torch.tensor([1.1]) 
w1 = torch.tensor([2.2], requires_grad=True) # set requires_grad to True to compute gradients
b = torch.tensor([0.0], requires_grad=True) # set requires_grad to True to compute gradients

z = x1 * w1 + b
a = torch.sigmoid(z)

loss = F.binary_cross_entropy(a, y)

grad_L_w1 = grad(loss,w1, retain_graph=True) # compute the gradient of the loss with respect to weight
grad_L_b = grad(loss,b, retain_graph=True) # compute the gradient of the loss with respect to bias

grad_L_w1, grad_L_b
```

- the retain_graph function is used so the cg stays inside of PyTorch's memory because it deletes the graph automatically by default
- this code shows how the grad function gets used manually. Using this is good or experimentation, debugging, and demonstrating concepts 
- however, pt offeres high-level tools to automate this. we can call .backward on the loss and pytorch will compute all of the gradients on the nodes in the graph and these gradients get stored in the tensors' .grad functions

```python
loss.backward() # compute the gradients of the loss with respect to all tensors with requires_grad=True

w1.grad, b.grad # access the gradients of w1 and b
```

#### Key Idea 

Pt's autograd component takes care of the calculus of the matter in chains of tensor operations using the .backward method. 

## Implementing multilayer NN
---
- We just looked at Pt's tensor and autograd components, but we going to shift to how Pt is a library that implements Deep NN's 
- torch.nn.Module provides functionality as it lets us buld and train models and encapsulating the layers and operations to keep track of a model's parameters
- the _init_ constructor defines the networks layers and how they interact with da forward method. 
- the foward method descides how da data passes through the networks and forms a cg
- we don't need to implement the backward method ourselves in a nn but it is used to compute the loss function w/ respect to da models parameters



```python
class NeuralNetwork(torch.nn.Module):
    def __init__(self, num_inputs, num_outputs): 
        super().__init__() # call the __init__ method of the parent class (torch.nn.Module)

        self.layers = torch.nn.Sequential(

            # First Layer
            torch.nn.Linear(num_inputs, 30), # first layer with num_inputs and 30 neurons
            torch.nn.ReLU(), # activation function

            # Second Layer
            torch.nn.Linear(30, 20), # second layer with 30 inputs and 20 neurons
            torch.nn.ReLU(), # activation function

            # Output layer
            torch.nn.Linear(20, num_outputs) # output layer with 20 inputs and num_outputs neurons
        )
    def forward(self, x): # define the forward pass of the network
        logits = self.layers(x) # pass the input through the layers
        return logits
    
model = NeuralNetwork(50, 3)
print(model)
```

The Sequential class we used wasn't required but it was useful cuz we wwanted to execute the layers in a specific order. \
After instantiating the self.layers = Sequential(...) in the _init_ we just have to call self.layers instead of each layer individually. 

```python
num_params = sum(
    p.numel() for p in model.parameters() if p.requires_grad
)
print(f"The model has {num_params} trainable parameters")
```

any parameter that requires_grad=True is a trainable parameter and gets updated in training
- the torch.nn.Linear layers contain the trainable parameters. 
- each linear layer multiplies the input w/ the weight matrix and adds the bias vector  <-- this is ref to as feedforward or a fully conected layer. 

```python
print(model.layers[0].weight) #this shows the weight matrix that is multiplied against the inputs 
```

```python
print(model.layers[0].weight.shape) #this shows the dimentions of the matrix since its so massive and doesn't get shown entirely
```

- you can do the same for the bias vector with model.layers[0].bias
- this 30x50 matrix holds the defult trainable entries for weight in torch.nn.Linear
- the numbers within the matrix usually differ because the model weights are initiallized randomly to break symetry in training. Otherwise, the nodes would just be doing the same operations over and over which won't allow them to learn the mappings from inputs to outputs. 

```python
torch.manual_seed(123) #manual_seed allows the random number intialization to be reproducable, kinda like a minecraft seed

model = NeuralNetwork(50,3)
print(model.layers[0].weight)
```

## Setting up efficient data loaders
- in this section we'll implement a dataset class that we will use to create a training and test dataset that will be used to create a data loader. 


#### Class Label Numbering 
Pt reqirues that a class label needs to start with 0 and that the largest class label value should not be greater that the # of output nodes - 1 (since index counting starts at 0). so, if we have class labels 0, 1, 2, 3, and 4, the nn would have 5 nodes.  

```python
from torch.utils.data import Dataset

#Creating ToyDataset by subclassing from Dataset parent class

import torch
import torch.nn.functional as F
from torch.autograd import grad

X_train = torch.tensor([
     [-1.2, 3.1],
    [-0.9, 2.9],
    [-0.5, 2.6],
    [2.3, -1.1],
    [2.7, -1.5]
])

y_train = torch.tensor([0,0,0,1,1])   # These are the training Datasets

X_test= torch.tensor([
    [-0.8,2.8],
    [2.6,-1.6],
])

y_test = torch.tensor([0,1])  # These are the test Datasets

class ToyDataset(Dataset):
    def __init__(self, X, y): # _init_ is the constructor that sets up the attributes that we use in the other methods. This could be file paths, file objects, database connectors and more. This just assigns the variables to the attributes but they really just placeholders for the tensor objects for now. 
        self.features = X
        self.labels = y 

    def __getitem__(self, index):
        one_x = self.features[index]
        one_y = self.labels[index]
        return one_x, one_y # the _getitem_ function just returns one item from the data set using the index. the index is supposed to be provided by the data loader. 

    def __len__(self): 
        return self.labels.shape[0] # the _len_ method gets the length of the dataset. it just uses the .shape attribute to return the number of rows in an array. 
    
train_ds = ToyDataset(X_train, y_train)
test_ds = ToyDataset(X_test, y_test)
```

```python
from torch.utils.data import DataLoader


torch.manual_seed(123)

train_loader = DataLoader(
    dataset=train_ds,
    batch_size=2,
    shuffle=True,
    num_workers=0
)
test_ds = ToyDataset(X_test, y_test)

test_loader = DataLoader(
    dataset=test_ds,
    batch_size=2,
    shuffle=False,
    num_workers=0
)

for idx, (x,y) in enumerate(train_loader):
    print(f"Batch {idx+1}:",x,y) 
```
